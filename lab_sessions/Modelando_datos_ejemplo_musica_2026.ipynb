# Modelando un dominio con RDF

## Paso 1: Preguntas sobre el alcance y la estructura del conocimiento.

*   ¿Cuáles serán las clases de mi dominio?
*   ¿Es posible que mi contexto inicial se expanda en un futuro? ¿Cómo afecta eso a mi modelo?
*   ¿Qué características voy a modelar sobre cada una de las entidades de mi dominio?
* ¿Cómo deberían relacionarse las entidades de mi dominio?
* Lo que he detectado como instancia en mi domino, ¿es en realidad (o puede actuar como) una clase?

Estas preguntas no son, en esencia, muy distintas de las preguntas que nos hacemos a la hora de comenzar a modelar una base de datos relacional. Pero debemos adaptar nuestras respuestas y manera de pensar al paradigma RDF.
En esta primera etapa, tendríamos que obtener un producto que nos permita comprender las topologías (las estructuras) que esperamos que puedan darse en nuestros datos.Tendremos también que hacernos preguntas de caracter más técnico, tales como ontologías/tecnologías a usar para almacenar y exponer la información objetivo. Pero esa parte debería venir después de tener una respuesta a las preguntas que se han mencionado.

### **Alcance y potenciales vías de expansión:**
Las clases mencionadas anteriormente parten de una noción abstracta sobre conceptos relacionados con música. Pero es necesario contextuar el grafo que pretendemos que desarrollar. ¿Será un grafo sobre teoría musical? Clases como nota musical, o subclases de las mismas (do, re, mi, fusa, semifusa, corchea, etc.) tendrían sentido. ¿Tenemos pensado publicar información referida a bandas de éxito? Entonces seguramente notas musicales quedan fuera de nuestro dominio, pero hemos de insistir más en cuestiones como miembros de banda, festivales, etc. Sea cual sea el dominio que modelemos, tendríamos que estar preparados para que potenciales expansiones del mismo no arruinen nuestros esquemas.

* ¿Y si se hace el biopic de una banda? el vocalista de Queen es Fredy Mercury. Rami Malek no es Freddy Mercury. Pero el vocalista de Queen en la pelicula Bohemian Rhapsody es Rami Malek (haciendo de Freddy Mercury). Tendríamos que asegurarnos de que toda esa información puede ser representada de manera no ambigua.
*  Reemplazos: Lauren Hart es la vocalista del grupo Arch Enemy. Pero la vocalista de Arch Enemy hace un año era Alissa White-Gluz. Y Angela Gossow antes que ella. Y Johan Liiva aun antes. Si adopto desde un inicio una visión estática de qué personas se vinculas a qué bandas con qué roles, cambios en las bandas me podrían causar datos amgiuos o me podrían forzar a borrar datos que, aunque obsoletos, pueden seguir siendo relevantes.
*  Casos especiales: Vamos a meter la banda Gorillaz al grafo. Sus miembros oficiales son 2-D, Noodle, Murdoc Niccals y Russel Hobbs. Pero resulta que estos miembros son dibujos animados. EL vocalista es 2-D, pero 2-D no existe. Su voz es puesta por otras personas. PersonaS, en plural. Para entrevistas (a día de hoy) Kevin Bishop. En canciones, Damon Albarn. ¿Puede mi esquema de datos representar esto? Incluso asumiendo que pueda: ¿qué debe hacer si alguien pregunta en SPARQL quien es el vocalista de Gorillaz?


### **Características sobre cada entidad:**

Aunque en esta primara fase solemos trabajar a un nivel más bien abstracto, en este caso ayuda centrarse en la observación de individuos (instancias) concretas. Por ejemplo, cojamos la Banda Arch Enemy que mencionamos antes. ¿Qué podemos decir sobre ella?
* nombre: Arch Enemy (¿podría cambiar de nombre?)
* Género musical: death metal melódico (¿podrían ser varios? ¿habría que destacar alguno sobre el resto?).
* Su año de fundación: 1995.
* Discos: War Eternal, Khaos Legions, etc (hay que modelar distintas re-ediciones?).
* Canciones (¿deberíamos asociarlas al disco o a la banda?): The world is yours, The eagle flies alone, etc.
* Miembros activos (y miembros pasados).


Cogemos otra tipo de entidad. Por ejemplo, festival: Resurrection Fest. ¿Qué podemos decir sobre él?
* Nombre
* Distintas ediciones del mismo (¿ha de actuar como clase? O una edición ha de vincularse a su festival con otra propiedad que no sea instanciación?).
* De cierta edición: bandas participantes, horarios, lugar, fecha, escenarios, comida/bebida, instalaciones... de todos estos elementos, cabe hacerse preguntas.


Moraleja: Aparecen muchas preguntas siempre que modelamos CUALQUIER entidad.


* Cuando una noción se vuelve compleja y escapa la sencillez de sujeto-predicado-objeto para ser representada de forma completa (ejemplo: vigencia de la vinculación d eun miembro a una banda) tenemos que tomar la decisión de adoptar estructuras más complejas o renunciar a cierto grado de expresividad.

* Cuando dos tipos de entidades tienen una relación (ejemplo: canciones y bandas) hemos de plantearnos tratar de no repetir información (¿asociamos canción a disco y disco a banda? ¿o enlace explícito (y potencialmente redundante/inconsistente entre banda y canción?) y plantearnos la dirección más natural de esta relación (la banda compone una canción o la canción es compuesta por la banda)).

* Cuando hay clases disfrazadas de instancia (ejemplo: El festival Resurrection Fest: ¿es instancia o clase?). Debemos ser lo más pulcros posible, pero sin olvidar que, en realidad, no hay une necesidad estricta de que cierta URI se comporte exclusivamente como clase o como instancia. Si no va a haber razonamiento lógico con los datos, la decisión debería ser guiada, sobre todo, tratando de hacer el modelo lo más natural que sea posible. Si va a haber razonamiento, deberíamos tener en cuanta nociones referidas a OWL punning y sus implicaciones.

* Tenemos que ser conscientes de cuál es el objetivo de nuestro grafo. ¿Queremos dar soporte a cierto caso de uso? ¿O queremos modelar con brillantez todo lo que se pueda modelar en mi contexto? A partir de cierto punto, modelar cierta parte del dominio puede volverse muy complejo. Ese esfuerzo debería llevarse a cabo sólo si es necesario.



```python
#Ejemplos de distintas representaciones, más o menos expresivas, complejas o correctas, para indicar miembros de la banda Arch Enemy

# Correcto, pero perdemos mucha información
tt1_1 = """
:arch_enemy :member :Lauren_Hart .
"""

# Tenemos más información, pero, sin un razonador, una consulta para traer los miembros de cierta banda podría complicarse
# (involucra a muchas propiedades)
ttl_2 =  """
:arch_enemy :lead_vocalist :Lauren_Hart .
"""

# Más información, pero el problema anterior se agrava. Riesgo de tener una explisión de propiedades pro combinatoria de posibilidades.
ttl_3 =  """
:arch_enemy :current_lead_vocalist :Lauren_Hart ;
            :past_lead_vocalist :Alissa_white_Gluz .
"""

# Tenemos, a priori, toda la información. El esquema además nos permitiría representar miembros presentes y pasados con facilidad, usando
# propiedades como start_year y end_year (si la hubiere). El precio a pagar es que nuestro esquema se ha vuelto complejo, menos natural
# de entender y las consultas SPARQL que necesitaremos para rescatar información serán más difíciles de escribir.
ttl_4 =  """
:arch_enemy :has_member :Laurent_Hart_member_Arch_enemy .

:Laurent_Hart_member_Arch_enemy   a           :member_node ;
                                  :member     :Lauren_Hart ;
                                  :role       :lead_vocalist ;
                                  :start_year 2026 .
"""

# Mismo esquema que el anterior, pero no hacemos algo "feo", que es decir que Arch Enemy (:arch_enemy) tiene como miembro (:has_member)
# una cosa que no representa a un miembro (eso sería :Lauren_Hart), sino un nodo con un montón de información asociada (:Laurent_Hart_member_Arch_enemy).
# Ahoa creamos un nodo anónimo que incluye la misma informacion.
ttl_4 =  """

_:Launrent_Arch_enemy_membership  a :person_band_membership ;
                                  :member     :Lauren_Hart ;
                                  :role       :lead_vocalist ;
                                  :start_year 2026 ;
                                  :band :Arch_Enemy .

"""


# Exactamente la misma información con la misma estructura que en el caso anterior, pero usando una sintaxis alternativa.
# Así la escritura es más compacta, pero ojo, perdemos la capacidad de referenciar desde nuestro propio grafo al nodo que
# antes se llamaba _:Launrent_Arch_enemy_membership. Con esta sintaxis no tiene ningún identificador, es aún "más anónimo"
ttl_5 = """

[  a :person_band_membership ;
   :member     :Lauren_Hart ;
   :role       :lead_vocalist ;
   :start_year 2026 ;
   :band :Arch_Enemy ;
]

"""

```

## Paso 2: Conexión con otras ontologías o grafos de conocimiento

Una vez tenemos claro el alcance y la estructura de lo que queremos representar, es hora de hacerse otras preguntas que tienen que ver con prevenir esfuerzos innecesarios e incorrectos por "tratar de reinventar la rueda" y con  seguir buenas praxis de web semántica.

Las preguntas son, esencialmente, las siguientes:

* Las entidades que pretendo representar en mi dominio, ¿están ya representadas en algún otro grafo de conocimiento? Si es así, ¿debo utilizar sus URIs (identificadores) en mi propio grafo?

* Las clases de mis entidades, ¿han sido definidas ya, aunque sea de manera aproximada, en otra ontología? Si es así, ¿debo utilizar esas clases en mis grafos?

* Las propiedades que unen conceptos en mi grafo, ¿ya han sido definidas en otras ontologías? Si lo han sido, son compatibles en dominio y rango con mi elección de clases? ¿Debería utilizarlas?

* Para todas aquellas entidades, clases o propiedades que deba crear porque no existen en otros grafos, o existen pero he decidido no utilizarlas: ¿cuál debería ser mi esquema de URIs?


### Explorando un dominio de ejemplo: Bandas musicales


Comencemos por lo esencial. Vamos a coger entidades representativas de mi dominio e investigar si existen en fuentes conocidas. Sin duda, el tipo de entidad principal de mi dominio es la idea de "Bandas Musicales". Como primer paso, debería investigar si existe una ontología sobre bandas musicales, o cuestiones relacionadas con la música pero que me permita modelar bandas con todo lo que necesito. De existir, cuan completa, importante o correcta parece ser.

Cosas que se pueden intentar en este paso:
* Probar a poner palabras clave en Linked Open Vocabularies (https://lov.linkeddata.es/dataset/lov/). Aparecen varias cosas prometedoras: music ontology, music, menciones a Dbpedia, rdau... al hacer click en unas cuantas, me saltan errores 404 y advertencias de seguridad por contenido no cifrado.

* Probar a buscar ontologías o grafos en la Linked Open Data Could (https://lod-cloud.net/) --> De nuevo, aparecen unas cuantas cosas prometedores, y al tirar del hilo encuentro muchos proyectos mal referenciados o muertos (bbc-music, musicbrainz, etc.).

* Probar a buscar entidades representativas en grafos de conocimiento relevantes (Wikidata, DBpedia, etc.). En este caso, encontramos muchas cosas que parecen utiles, a nivel de clases y a nivel de individuos. Aquí tenemos dos vías de acción: por un lado, podríamos adoptar las URIs de cada banda/concepto en estos grafos como "propias" al definir nuestro datos. NO crearíamos ninguna URI nueva de ninguna entidad de nuestro dominio. Por otro lado, podríamos mantener nuestras propias URIs y enlazar con la de estos grafos a través de declaraciones. Habría que determinar el nivel de rotundidad o implicación logica que esos links tendrían:
  * Máxima a todos los niveles --> owl:sameAs/owlEquivalent class.
  * Misma entidad, pero sin equivalencia ontológica --> skos:exactMatch.
  * Correspondencia aproximada, podría haber ciertos matices --> skos:closeMatch.
  * Ninguna implicación lógica, pero "sabemos" nosotros humanos que esto quiere decir que dos URIs representan lo mismo y puede ser útil por RDFa --> schema:sameAs.
* Probar en buscadares web palabras clave + ontología / RDF / grafo conocimiento. Generalmente, dará mejores resultados hacer estas búsquedas en ingles.

En base a los elementos encontrados, debemos de ser capaces de decidir cuál es la mejor estrategia posible para representar nuestros datos.

A nivel de instancia, en caso de haber encontrado definiciones de identificadores existentes para tus entidades, usar esos mismos identificadores en tu grafo no es estrictametne necesario desde el punto de vista técnico, pero sí puede ser conveniente. Casi siempre, hay pros y contras de adoptar una u otra decisión en este aspecto.

#### Caso de Wikidata

Las URIs de Wikidata son "opacas". En lugar de tener URIs con nombres reconocibles, pasaríamos a tener identificadorer terminados en Qxxxxx. Eso podría no ser conveniente para la usabilidad/legibilidad de tus datos.

Por otro laod, si queremos usar sólo identificadores existentes en Wikidata ¿Podemos estar seguros de que si TODAS las entidades que necesitamso estarán en Wikidata? Es muy probable que sí, puesto que incluso nosotros mismos podemos provocar ese cambio editando la fuente. Eso sí, aunque sea poco probable, estaríamos expuestos a situaciones de bandalismo de datos en Wikidata, puesto que cualquier podría editar los items que nosotros mismos editamos. Con todo, sería nrazonable pensar que podemos asumir que esa lista puede actualizarse de acuerdo a nuestras necesidades.

Pero ojo, eso sirve para las entidades. Si necesitasemos una propiedad que no existe, no tendríamso más remedio que iniciar una discusión comunitaria, y cruzar los dedos porque se resuelva pronto y de manera positiva.

Si en lugar de usar la URIs de Wikidata desarrollasemos nuestras propias URIs y las enlazasemos con Qs de Wikidata mediante tirpletas owl:sameAs/skos:exactMatch, podríamos mantener URIs legibles, mantenernos enlazados a una fuente tan relevante como Wikidata y podríamos seguir creando URIs de nuevas entidades y propiedades sin depender de que estos estén o no en Wikidata. En el momento en que entren, añadiríamos una tripleta para enlazar nuestra URI y la de Wikidata. El único contra de esta opción es que tendríamos que añadir esas tripletas owl:sameAs/skos:exactMatch a nuestros datos --> dataset más grande para contener la misma información y las consultas federadas se volverán algo más complejas.


#### Caso de DBpedia

A efectos prácticos de tomar esta decisión respecto a las URIs, Wikidata y DBpedia tienen en común ser proyectos de reconocido prestigio y muy centrales. Merece la pena enlazarse con ellos. Tienen estos dos rasgos diferentes de especial relevancia para la cuestión: las URIs de DBpedia no son opacas y tampoco las actualiza la comunidad (o no directamente al menos).

Esto quiere decir que si adoptamos las URIs de DBpedia para deescribir bandas de música en nuestro grafo, ganaríamos en legibilidad. Pero, eso sí, perderíamos en capacidad de reacción. Si aparecen nuevas bandas y quisiésemos meter datos sobre ellas, no tenemos control en absoluto respecto a cuando (o si quiera si ocurrirá) serán introducidas URIs para estos elementos en DBpedia. Para evitar ese problema, podríamos hacer mezcla de URIs de DBpedia con URIs propias allá donde haya un nuevo elemento no incluido en la fuente. Pero esto queda "feo".

Las ventajas e inconvenientes de incluir sentencias owl:SameAs enlazando URIs propias de nuestro espacio de nombre y URIs de DBpedias son las mismas que en el caso de Wikidata.

#### Caso de otras ontologías


Si la ontología es correcta y completa para nuestros fines, hay que poner en una balanza el trabajo que nos ahorra usarla a nivel de modelado de clases y propiedades o incluso de aprovechamiento de información disponible versus el riesgo de que deje de mantenerse. Y ya hemos visto que hay un montón de ontologías relacionadas con la música que parecen proyectos muertos.

Ejemplo: supón que das por bueno todo lo que ahí está escrito, lo que te ahorra hacer un montón de trabajo de modelado y de recopilación de información. Si tu trabajo original requiere hablar de bandas, miembros, roles, instrumentos y conciertos, puede que la mayor parte de esa información ya exista. La terea podría verse reducida a, por ejemplo, anotar conciertos.

Tendría todo el sentido en ese caso adoptar las URIs de ese grafo y simplemente añadir información a los datos existentes. No se trataría de copiar y pegar esos datos existentes, sino en confiar en su veracidad y enlazarlos desde tus propios datos. Ahora bien: ¿y si hubiese que corregir los datos? No tendríamos en absoluto control sobre las ediciones ni ninguna garantía de que estas vayan a llevarse a cabo.

En un escenario real, donde pretendemos publicar datos importantes para X objetivo, una vía de acción válida sería contactar personalmente con la persona o entidad que mantiene la ontología. Conocer sus intenciones y posibilidades respecto a la estabilidad de los datos, su mantenimiento en caso necesario o incluso la posibilidad de pasar a formar parte del grupo de personas con capacidad para mantener la aontología.


### Esquema de URIs

Decidas lo que decidas sobre qué datos reusar, muy probablemente vas a generar nuevas URIs. Tengas o no intención de dejar públicos tus datos, deberías siempre considerar tener un esquema de URIs consistente, bien organizado, centrado en el dominio y tan natural y sencillo como sea posible.

En primer lugar, el inicio de la URI, a no ser que estemos usando example.org, debería hacer referencia a algún dominio sobre el que tuvieses control y que sea pertinente para el dominio modelado.

Por ejemplo, yo, si tengo que publicar datos consecuencia de mi actividad investigadora, lo haría en el dominio http://weso.es/, correpondiente a mi grupo de investigación, sobre el que el grupo tiene pleno control. Si estuviese trabajando para otra empresa, usaría el dominio de esa empresa.
En vuestro caso, en el contexto de esta asignatura, podría tener sentido, por ejemplo, publicar datos en el dominio de uniovi, a pesar de que no está bajo vuestro control, pero con más niveles de profundidad. Por ejemplo, tendría sentido hacer algo como esto:







```python
ttl_8 = """

PREFIX miw : <http://uniovi.es/miw/ws/tu_UO/tu_tema#>

miw:Arch_Enemy a miw:Musical_Band .
"""
# miw por master de ingeniería web, ws por web semántica, UO paa que nadie de
# esta asignatura y de uniovi te pise el dominio, más alguna palabra clave descriptiva.
```

Para X aplicación, para Y análisis. No todo tiene que ser LOD cloud impecable o ser compatible con razonadores OWL.
